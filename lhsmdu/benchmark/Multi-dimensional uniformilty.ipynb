```python
import lhsmdu
import matplotlib.pyplot as plt
import numpy as np
```

```python
dimensions = [2**x for x in xrange(1,4)]
samples = [4**x for x in xrange(1,4)]
```

```python
def simpleaxis(axes, every=False):
    if not isinstance(axes, (list, np.ndarray)):
        axes = [axes]
    for ax in axes:
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        if every:
            ax.spines['bottom'].set_visible(False)
            ax.spines['left'].set_visible(False)
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()
        ax.set_title('')
```

```python
for i, sample in enumerate(samples):
    for j, dims in enumerate(dimensions):
        print("{} samples, {} dims".format(sample, dims))
        a = lhsmdu.sample(dims, sample)
        b = np.random.random(size=(dims, sample))
        fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True, figsize=(4,2))
        ax[0].scatter(np.ravel(a[0]), np.ravel(a[1]), c='orange')
        ax[1].scatter(np.ravel(b[0]), np.ravel(b[1]), c='k')
        simpleaxis(ax)
        ax[0].set_title("LHSMDU")
        ax[1].set_title("MCS")
        plt.show()        
```
